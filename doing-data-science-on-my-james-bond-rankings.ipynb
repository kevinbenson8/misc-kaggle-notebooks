```python
# create dataframe
df = pd.read_csv('../input/jamesbond/jamesbond.csv')
columns = ['Year', 'Movie', 'Bond', 'Director', 'Film_Length', 'Avg_User_IMDB']
df = df[columns]
```

```python
# add 'No Time to Die' do dataset
nttd = {'Year': 2021,
        'Movie': 'No Time to Die',
        'Bond' : 'Daniel Craig',
        'Director' : 'Cary Joji Fukunaga',
        'Film_Length' : 163,
        'Avg_User_IMDB': 7.3,
       }

df = df.append(pd.DataFrame(nttd, index=[0]), ignore_index=True)
```

```python
# Merge with my movie rankings
scores = pd.read_csv("../input/bondrankings/Bond_rankings.csv")
df = df.merge(scores, how='inner')
```

```python
df.tail()
```

```python
# sort by total score
df[['Year', 'Movie', 'Bond', 'Total score']].sort_values(by='Total score', ascending=False)
```

```python
# compare my scores to imdb scores
df['score_difference'] = df['Total score']/3 - df['Avg_User_IMDB']
df[['Year', 'Movie', 'Bond', 'score_difference']].sort_values(by='score_difference', ascending=False)
```

```python
# group by actor to get average score for each bond
df.groupby('Bond')['Total score'].mean().sort_values(ascending=False)
```

```python
# same thing but with Bond Masculinity meter instead of total score
df.groupby('Bond')['Bond masculinity meter'].mean().sort_values(ascending=False)
```

```python
# plot bar graph to get an idea how the scores of each movie are comprised

import plotly.express as px

fig = px.bar(df,
             x = ['Movie score', 'Bond masculinity meter', 'Bond girl gauge', 'Evil index', 'Q points'],
             y = 'Movie',
             title = 'Bond Movie Scores',
             orientation='h',
             width=800, height=1200
             )
fig.update_layout(legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="left",
    x=0.67,
    title="",
    font=dict(size= 10)),
    xaxis_title=None,
    yaxis_title=None
)
fig.update_yaxes(tickfont_family="Arial Black")
fig.show()
```

```python
# line plot to see if there is a trend in movie scores over time

fig = px.line(df,
             x = 'Year',
             y = ['Movie score', 'Avg_User_IMDB'],
             title = 'Bond Movie Scores by Year',
             )
fig.show()
```
